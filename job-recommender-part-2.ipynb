```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Imports the TF-IDF vectorizer from scikit-learn for text feature extraction
```

```python
# Creates a TF-IDF vectorizer object, removing common English stop words
# Tạo đối tượng vectơ hóa TF-IDF, loại bỏ các từ dừng tiếng Anh phổ biến
tdif=TfidfVectorizer(stop_words='english')

# Fills missing values in the job description column with empty strings (potentially a better option than dropping rows).
df1['jobdescription']=df1['jobdescription'].fillna('')

# Applies TF-IDF vectorization to the job descriptions, converting them into a numerical representation.
tdif_matrix=tdif.fit_transform(df1['jobdescription'])
# Shows the dimensions of the TF-IDF matrix (number of documents, number of features).
tdif_matrix.shape
```

4. Cosine Similarity Calculation

```python
# Nhập hàm kernel tuyến tính để tính toán độ tương đồng cosine.
from sklearn.metrics.pairwise import linear_kernel

# Tính toán ma trận độ tương đồng cosine giữa tất cả các mô tả công việc. Độ tương đồng cosine đo lường mức độ giống nhau của hai tài liệu dựa trên vectơ TF-IDF của chúng.
cosine_sim=linear_kernel(tdif_matrix,tdif_matrix)

# Tạo một Series ánh xạ các tiêu đề công việc duy nhất với các chỉ mục tương ứng của chúng trong DataFrame (để truy xuất hiệu quả).
indices=pd.Series(df1.index, index=df1['jobtitle']).drop_duplicates()
```

```python
# nhận tiêu đề công việc làm đầu vào và trả về các công việc được đề xuất.
def get_recommendation(title,cosine_sim=cosine_sim):
    # Lấy chỉ mục của tiêu đề công việc đầu vào trong DataFrame.
    idx=indices[title]
    # Tạo danh sách các bộ nhớ đệm, trong đó mỗi bộ nhớ đệm chứa chỉ mục của công việc tương tự và điểm số độ tương đồng cosine của nó với công việc đầu vào
    sim_scores=list(enumerate(cosine_sim[idx]))
    # Sắp xếp các công việc tương tự theo điểm số độ tương đồng cosine của chúng theo thứ tự giảm dần (giống nhau nhất trước).
    sim_scores=sorted(sim_scores,key=lambda X: X[1], reverse=True)
    # Chọn 15 cv đầu . ko lấy cv đầu vào ở chỉ mục 0
    sim_scores=sim_scores[1:16]
    tech_indices=[i[0] for i in sim_scores]
    # Sử dụng vòng lặp comprehension để duyệt qua danh sách sim_scores. Danh sách này chứa các bộ nhớ đệm, mỗi bộ nhớ đệm bao gồm chỉ mục công việc và điểm số độ tương đồng cosine tương ứng.
    # Trong mỗi vòng lặp, mã trích xuất phần tử đầu tiên (chỉ mục 0) của bộ nhớ đệm hiện tại. Đây chính là chỉ mục của một công việc tương tự.
    # Chỉ mục này được thêm vào danh sách tech_indices.
    return df1['jobtitle'].iloc[tech_indices]
```
